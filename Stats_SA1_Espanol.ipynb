```python
import pandas as pd
import numpy as np
from scipy import stats

#Data Input
males = [12, 4, 11, 13, 11, 7, 9, 10, 10, 7, 7, 12, 6, 9, 15, 10, 11, 12, 7, 8,
         8, 9, 11, 10, 9, 10, 9, 9, 7, 9, 11, 7, 10, 10, 11, 9, 12, 12, 8, 13,
         9, 10, 8, 11, 10, 13, 13, 9, 10, 13]

females = [11, 9, 7, 10, 9, 10, 10, 7, 9, 10, 11, 8, 9, 6, 11, 10, 7, 9, 12, 14,
           11, 12, 12, 8, 12, 12, 9, 10, 11, 7, 12, 7, 9, 8, 11, 10, 8, 13, 8, 10,
           9, 9, 9, 11, 9, 9, 8, 9, 12, 11]

df_all = pd.concat([pd.DataFrame({'Hours': males, 'Gender': 'Male'}),
                    pd.DataFrame({'Hours': females, 'Gender': 'Female'})])

# Statistical Function
def calculate_formatted_stats(data):
    # Calculate basic descriptives and deciles
    percentiles = [0.1, 0.2, 0.25, 0.3, 0.4, 0.5, 0.6, 0.7, 0.75, 0.8, 0.9]
    s = data.describe(percentiles=percentiles)

    s.rename({'count': 'sample_size (n)'}, inplace=True)
    s['variance'] = data.var()
    s['skewness'] = data.skew()
    s['kurtosis'] = data.kurtosis()

    # 95% Confidence Interval
    mean, sem = data.mean(), stats.sem(data)
    ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)
    s['95% CI Lower'] = ci[0]
    s['95% CI Upper'] = ci[1]

    #put sample size at the top
    items = list(s.index)
    items.insert(0, items.pop(items.index('sample_size (n)')))
    return s.reindex(items)

# Q1
print("--- I.1: ENTIRE DATASET SUMMARY ---")
overall = calculate_formatted_stats(df_all['Hours'])
# Use .to_string() to remove the "Name/dtype" line at the bottom
print(overall.to_string())

print("\n" + "="*40 + "\n")

# Q2
print("--- I.2: SUMMARY BY GENDER ---")
gender_grouped = df_all.groupby('Gender')['Hours'].apply(calculate_formatted_stats).unstack().T
print(gender_grouped.to_string())
```

# Part III: Mathematical Proof
**Prove that $m'_4 = m_4 + 4hm_3 + 6h^2m_2 + h^4$, given $h = m'_1$**

Define $m'_4 = E[X^4]$ (4th raw moment)

$m_k = E[(X - \mu)^k]$ (central moment)

$h = \mu$ (mean)

Expansion

Express $X$ as $(X - \mu) + \mu$. Substitute $h$ for $\mu$

$$m'_4 = E[((X - \mu) + h)^4]$$

Use Binomial Expansion

Let $a = (X - \mu)$ and $b = h$

$(a + b)^4 = a^4 + 4a^3b + 6a^2b^2 + 4ab^3 + b^4$

Then, $$m'_4 = E[(X - \mu)^4 + 4h(X - \mu)^3 + 6h^2(X - \mu)^2 + 4h^3(X - \mu) + h^4]$$

Distribute the Expectation $E[\cdot]$ to each term:

$$m'_4 = E[(X - \mu)^4] + 4hE[(X - \mu)^3] + 6h^2E[(X - \mu)^2] + 4h^3E[(X - \mu)] + E[h^4]$$

Simplify

$E[(X - \mu)^4] = m_4$$E[(X - \mu)^3] = m_3$$E[(X - \mu)^2] = m_2$$E[(X - \mu)] = 0$

(first central moment is always zero)

$E[h^4] = h^4$ (The expectation of a constant is the constant)

Final Result

Substitute back into the equation:

$$m'_4 = m_4 + 4h(m_3) + 6h^2(m_2) + 4h^3(0) + h^4$$

Therefore, $$m'_4 = m_4 + 4hm_3 + 6h^2m_2 + h^4$$
