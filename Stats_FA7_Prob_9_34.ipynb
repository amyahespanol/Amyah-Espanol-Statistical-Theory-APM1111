```python
import numpy as np
from scipy import stats

males = np.array([
    85, 53, 100, 49, 65, 60, 51, 61, 83, 65,
    55, 99, 56, 55, 55, 90, 72, 62, 69, 59,
    49, 72, 58, 60, 68, 90, 74, 85, 80, 77,
    62, 65, 81, 55, 71, 78, 49, 78, 80, 75,
    53, 82, 109, 87, 78, 72, 104, 70, 31, 50
])

females = np.array([
    64, 93, 82, 71, 77, 64, 60, 75, 87, 60,
    61, 84, 91, 61, 85, 105, 90, 59, 86, 62,
    71, 99, 98, 54, 94, 98, 61, 108, 79, 50,
    66, 74, 60, 90, 95, 81, 86, 65, 86, 81,
    92, 77, 82, 86, 79, 91, 93, 63, 93, 53
])

#Calculate stats
mean_diff = np.mean(males) - np.mean(females)
n_m, n_f = len(males), len(females)
var_m, var_f = np.var(males, ddof=1), np.var(females, ddof=1)

#Standard Error
se_diff = np.sqrt((var_m / n_m) + (var_f / n_f))

#85% Confidence Limit
#0.85 + (0.15/2) = 0.925
z_85 = stats.norm.ppf(0.925)
margin_of_error = z_85 * se_diff

ci_85 = (mean_diff - margin_of_error, mean_diff + margin_of_error)

print(f"Mean Difference: {mean_diff:.2f} years")
print(f"Standard Error: {se_diff:.4f}")
print(f"85% Confidence Interval: {ci_85}")
print(f"85% Confidence Interval: ({ci_85[0]:.2f}, {ci_85[1]:.2f})")
```
